```python


def clean_domaine(val):
    if pd.isna(val):
        return 'Autre'
    
    v = val.strip().lower()
    
    if re.search(r'data|ia|intelligence artificielle|machine learning|data science|data engineer|data scientist|data & ai|data/ia|data et ia|data ia', v):
        return 'Data / IA'
    elif re.search(r'info|informatique|maths.info|dev|développement|numérique', v):
        return 'Informatique'
    elif re.search(r'marketing|communication|com\b', v):
        return 'Marketing & Com'
    elif re.search(r'finance|banque|assurance|compta', v):
        return 'Finance / Banque'
    elif re.search(r'droit|juridique', v):
        return 'Droit'
    elif re.search(r'maths|statistique|sciences', v):
        return 'Maths / Stats'
    elif re.search(r'industri|automobile|ingénierie', v):
        return 'Industrie'
    elif re.search(r'langue|lettres|langues', v):
        return 'Langues'
    elif re.search(r'conseil|consulting|crm', v):
        return 'Conseil'
    else:
        return 'Autre'

df['domaine_clean'] = df['domaine'].apply(clean_domaine)
print(df['domaine_clean'].value_counts())
```

```python
def clean_nb_cand(val):
    if pd.isna(val): return np.nan
    v = str(val).strip().lower()
    if any(x in v for x in ['non', 'aucun']): return 0
    if any(x in v for x in ['centaine', '100aine', 'une cinquantaine']): return 100
    if 'beaucoup' in v: return 150
    if '200/300' in v: return 250
    if '120-130' in v: return 125
    return pd.to_numeric(v, errors='coerce')

df['nb_candidatures_clean'] = df['nb_candidatures'].apply(clean_nb_cand)
print(df['nb_candidatures_clean'].describe())
```

```python
def clean_feature(val):
    if pd.isna(val): return 'Autre'
    v = str(val).strip().lower()
    if any(x in v for x in ['entretien', 'interview']): return 'Simulation entretien'
    if any(x in v for x in ['cv', 'curriculum']): return 'CV optimisé'
    if any(x in v for x in ['profil', 'métier', 'ciblage']): return 'Analyse profil → métiers'
    if any(x in v for x in ['apprentissage', 'plan']): return 'Plan apprentissage'
    if any(x in v for x in ['offre', 'adapt']): return 'Adaptation offres'
    return 'Autre'

df['feature_clean'] = df['feature_utile'].apply(clean_feature)
print(df['feature_clean'].value_counts())
```

```python
def clean_wtp(val):
    if pd.isna(val): return 'Autre'
    v = str(val).strip().lower()
    if any(x in v for x in ['oui', 'yes', '✅', 'ouiii']): return 'Oui'
    if any(x in v for x in ['peut-être', 'ptet', 'possible']): return 'Peut-être'
    if 'non' in v: return 'Non'
    return 'Autre'

df['wtp_clean'] = df['willingness_to_pay'].apply(clean_wtp)
print(df['wtp_clean'].value_counts())
```

```python
for col in ['stress_pro', 'probleme_unique', 'manque_solutions']:
    print(f'\n--- {col} ---')
    for v in df[col].dropna().tolist():
        print('-', v)
```

```python
def sc(s): return {str(k): int(v) for k, v in s.value_counts().items()}
def sm(s): return round(float(pd.to_numeric(s, errors='coerce').mean()), 2)

stats = {
    'total': len(df),

    # Profil
    'age': sc(df['age']),
    'situation': sc(df['situation'].str.split(',').explode().str.strip()),
    'niveau_etudes': sc(df['niveau_etudes']),
    'domaine': sc(df['domaine_clean']),

    # Problème
    'clarte_metier': sc(df['clarte_metier'].str.split(',').explode().str.strip()),
    'score_perdu_moyen': sm(df['score_perdu']),
    'score_perdu_distrib': sc(df['score_perdu']),
    'blocage': sc(df['blocage'].str.split(',').explode().str.strip()),
    'test_orientation': sc(df['test_orientation']),

    # Compétences
    'comp_cv': sm(df['comp_cv']),
    'comp_lm': sm(df['comp_lm']),
    'comp_adapter_cv': sm(df['comp_adapter_cv']),
    'comp_entretien': sm(df['comp_entretien']),
    'connaissance_competences': sm(df['connaissance_competences']),
    'connaissance_salaires': sm(df['connaissance_salaires']),
    'connaissance_debouches': sm(df['connaissance_debouches']),
    'connaissance_evolutions': sm(df['connaissance_evolutions']),

    # Candidatures
    'a_postule': sc(df['a_postule']),
    'nb_candidatures_moyen': sm(df['nb_candidatures_clean']),
    'taux_reponse': sc(df['taux_reponse']),

    # Validation marché
    'interet_app': sc(df['interet_app']),
    'feature_utile': sc(df['feature_clean']),
    'willingness_to_pay': sc(df['wtp_clean']),
    'utiliserait_app': sc(df['utiliserait_app']),

    # Verbatims
    'stress_pro': df['stress_pro'].dropna().tolist(),
    'probleme_unique': df['probleme_unique'].dropna().tolist(),
    'manque_solutions': df['manque_solutions'].dropna().tolist(),
}

with open('career_app_stats.json', 'w', encoding='utf-8') as f:
    json.dump(stats, f, ensure_ascii=False, indent=2)

print(' Done')
```
